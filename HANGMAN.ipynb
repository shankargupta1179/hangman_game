```python
import random
class game():
    
    def __init__(self, ans_word, hints):
        self.ans_word=ans_word
        self.max_guesses=len(self.ans_word)+2
        self.guess_count=0
        self.hints=hints
        self.totalscore=0
        self.d={}
        self.lowered=self.ans_word.lower()
        for i in set(self.lowered):
            self.d[i]=0
        for i in self.lowered:
            self.d[i]+=1
        print("HERE'S YOUR HINT\n",hints)
        print(" _ "*len(ans_word))
        
    def choosestatement(self,win=0):
        self.win_statements=["Woah! Nailed it.", "Righty right.", "Smart move!"]
        self.loss_statements=["Better luck next time", "Tera bhi din ayga", "Wrong guess Kid! Hahahahaha."]
        if win==1:
            return random.choice(self.win_statements)+"\nYou've {} chances left".format(self.max_guesses-self.guess_count)
        else:
            return random.choice(self.loss_statements)+"\nYou've {} chances left".format(self.max_guesses-self.guess_count)

    def display(self, guessed_letters):
        l=[]
        for char in self.ans_word:
            if char.lower() in guessed_letters:
                l.append(char)
                
            else:
                l.append("_")
        s=' '.join(l)        
        print("According to your guesses till now ",s,"\n")
        return "".join(l)
                        
                
    def validate(self,score=0):
        
        guessed_letters=[]
        while self.guess_count<self.max_guesses: 
            self.guess_count+=1
            inputchar=input("Chote nhi hote hathi k paer is the, input kro guessed char is the\n")
            right_guess=0
            x=inputchar.lower()
            if x in self.d and self.d[x]:
                right_guess=1
                print(self.choosestatement(right_guess))
                guessed_letters.append(x)
                self.d[x]-=1
                
                
                
            else:
                print(self.choosestatement(right_guess))
                
            word_till=self.display(guessed_letters)
            
            if word_till==self.ans_word:
                print("Hurray! You win")
                score+=1
                print("\n The correct word was", self.ans_word)   
                break
        else:
            
            print("Game over.\n The correct word was", self.ans_word)
        return score
        
    def write_file(self, score, username="user1"):
        with open ("scorefile.txt", mode="a") as f:
            f.write(username+":"+str(score)+"\n")
                    
        
        
    
```
